```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

driver = webdriver.Chrome(ChromeDriverManager().install())
import pandas as pd
import time 
import os
```

```python
# driver.get method() will navigate to a page given by the URL address
driver.get("https://www.bbc.com/news/topics/cldy2dmy748t")
```

```python
# driver.find_elements(By.XPATH,' ') will extract multiple texts from the website
headline1 = driver.find_elements(By.XPATH,'//p[@class="ssrcss-17zglt8-PromoHeadline e1f5wbog4"]/span[@aria-hidden="false"]')
headline1[0].text
```

```python
# driver.find_elements(By.XPATH,' ') will extract multiple texts from the website
headline2 = driver.find_elements(By.XPATH,'//p[@class="ssrcss-17zglt8-PromoHeadline e1f5wbog4"]/span[2]')
headline2[0].text
```

```python
pagesToGet= 31
HeadlineList1_Crime = []
HeadlineList2_Crime = []
HeadlineList_Crime = []

for page in range(1,pagesToGet+1):
    print('processing page :', page)
    url = 'https://www.bbc.com/news/topics/cldy2dmy748t?page='+str(page)
    print(url)
    driver.get(url)
  
    headline1 = driver.find_elements(By.XPATH,'//p[@class="ssrcss-17zglt8-PromoHeadline e1f5wbog4"]/span[@aria-hidden="false"]')
    headline2 = driver.find_elements(By.XPATH,'//p[@class="ssrcss-17zglt8-PromoHeadline e1f5wbog4"]/span[2]')
    
    for i in range(len(headline1)):
        Headline = headline1[i].text.strip()
        HeadlineList1_Crime.append(Headline)
        print(headline1[i].text)

        
    for i in range(len(headline2)):
        Headline = headline2[i].text.strip()
        HeadlineList2_Crime.append(Headline)
        print(headline2[i].text)

   
    print('\n'+'\n'+ 'Total '+ str(len(HeadlineList1_Crime + HeadlineList2_Crime)) +' headlines.'+' till page no. ' + str(page) +'\n'+'\n')
    
    time.sleep(4)

HeadlineList_Crime =  HeadlineList1_Crime + HeadlineList2_Crime
print(len(HeadlineList_Crime))
```

```python
pagesToGet= 42
HeadlineList1_Politics = []
HeadlineList2_Politics = []
HeadlineList_Politics = []

for page in range(1,pagesToGet+1):
    print('processing page :', page)
    url = 'https://www.bbc.com/news/topics/cwnpxwzd269t?page='+str(page)
    print(url)
    driver.get(url)
  
    headline1 = driver.find_elements(By.XPATH,'//p[@class="ssrcss-17zglt8-PromoHeadline e1f5wbog4"]/span[@aria-hidden="false"]')
    headline2 = driver.find_elements(By.XPATH,'//p[@class="ssrcss-17zglt8-PromoHeadline e1f5wbog4"]/span[2]')
    
    for i in range(len(headline1)):
        Headline = headline1[i].text.strip()
        HeadlineList1_Politics.append(Headline)
        print(headline1[i].text)

        
    for i in range(len(headline2)):
        Headline = headline2[i].text.strip()
        HeadlineList2_Politics.append(Headline)
        print(headline2[i].text)

   
    print('\n'+'\n'+'Total '+str(len(HeadlineList1_Politics+HeadlineList2_Politics)) + ' headlines.'+ ' till page no. '+ str(page) + '\n'+ '\n')
    
    time.sleep(4)
    
HeadlineList_Politics =  HeadlineList1_Politics + HeadlineList2_Politics
print(len(HeadlineList_Politics))
```

```python
pagesToGet= 2
HeadlineList1_Finance = []
HeadlineList2_Finance = []
HeadlineList_Finance = []

for page in range(1,pagesToGet+1):
    print('processing page :', page)
    url = 'https://www.bbc.com/news/topics/c4qzyd68613t?page='+str(page)
    print(url)
    driver.get(url)
 
    headline1 = driver.find_elements(By.XPATH,'//p[@class="ssrcss-17zglt8-PromoHeadline e1f5wbog4"]/span[@aria-hidden="false"]')
    headline2 = driver.find_elements(By.XPATH,'//p[@class="ssrcss-17zglt8-PromoHeadline e1f5wbog4"]/span[2]')
    
    for i in range(len(headline1)):
        Headline = headline1[i].text.strip()
        HeadlineList1_Finance.append(Headline)
        print(headline1[i].text)

        
    for i in range(len(headline2)):
        Headline = headline2[i].text.strip()
        HeadlineList2_Finance.append(Headline)
        print(headline2[i].text)

   
    print('\n'+'\n'+'Total '+str(len(HeadlineList1_Finance+HeadlineList2_Finance)) + ' headlines.'+ ' till page no. '+ str(page) + '\n'+ '\n')
    
    time.sleep(4)
    
HeadlineList_Finance =  HeadlineList1_Finance + HeadlineList2_Finance
print(len(HeadlineList_Finance))
```

```python
Headline_df_Crime = pd.DataFrame({'Article' : HeadlineList_Crime ,'Label' : ['Crime']*len(HeadlineList_Crime) })
print(Headline_df_Crime.head())
print(Headline_df_Crime.info())
```

```python
Headline_df_Politics = pd.DataFrame({'Article' : HeadlineList_Politics ,'Label' : ['Politics']*len(HeadlineList_Politics) })
print(Headline_df_Politics.head())
print(Headline_df_Politics.info())
```

```python
Headline_df_Finance = pd.DataFrame({'Article' : HeadlineList_Finance ,'Label' : ['Finance']*len(HeadlineList_Finance) })
print(Headline_df_Finance.head())
print(Headline_df_Finance.info())
```

```python
Headline_df = pd.concat([Headline_df_Crime,Headline_df_Politics,Headline_df_Finance])
print(Headline_df.head())
print(Headline_df.info())
```
