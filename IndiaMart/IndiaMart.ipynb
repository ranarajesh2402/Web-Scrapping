```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(ChromeDriverManager().install())
import pandas as pd
import time
import os
```

#### Used the link of IndiaMart for Ceiling Titles 

```python
# driver.get method() will navigate to a page given by the URL address
driver.get("https://dir.indiamart.com/search.mp?ss=ceiling+tiles&src=as-popular%3Akwd%3Dceilingtiles%3Apos%3D1%3Acat%3D-2%3Amcat%3D-2")
```

### NOTE: To get the phone number of the retailer you need to login by phone no. and OTP. Then click on "See More Result" again and again to get more data at once.

```python
from selenium.webdriver.common.by import By

storeDetails = driver.find_elements(By.XPATH, '//div[@class="dib pr to-txt-gn"]/span[@class="elps elps1"]')
names = driver.find_elements(By.XPATH, '//h4[@class="lcname tac"]/a[@class="clr3 fs12 fwn rsrc"]/span')

#To get the phonenumber of the retailer you need to login and OTP. Then click on See More Result
contactList = driver.find_elements(By.XPATH, '//span[@class="pns_h duet fwb"]')

nameList = []
addressList = []
numbersList = []

for i in range(len(names)):
    
    name = names[i].text
    address = storeDetails[i].text
    contact = contactList[i].text
    
    nameList.append(name)
    addressList.append(address)
    numbersList.append(contact)


    
# intialise data of lists.
data = {'Company Name':nameList,
        'Address': addressList,
        'Phone':numbersList
       }

# Create DataFrame
df = pd.DataFrame(data)
df
```

```python
df.reset_index(drop=True)
```

```python
df.to_csv('IndiaMart_ceiling-tile.csv',index=False)
```
